```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import pandas as pd
import requests
import re
```

```python
driver = webdriver.Chrome(executable_path='chromedriver.exe')
driver.get('https://my.clevelandclinic.org/staff?q=a&dFR[employedDescription][0]=Cleveland%20Clinic%20employed%20staff')
time.sleep(2)

while True:
    try:
        driver.find_element(By.XPATH, '//*[@id="pagination"]/a').click()
        time.sleep(1)
    except:
        break
    
soup = BeautifulSoup(driver.page_source, 'html.parser')
url = []
base_url = 'https://my.clevelandclinic.org'
for x in soup.find_all("h2"):
    try:
        url.append(base_url+x.find_all('a', href=True)[0]['href'])
    except:
        pass
```

```python
# Specify the file name
file_name = "urls.txt"

# Open the file in write mode
with open(file_name, "w") as file:
    # Iterate through the list and write each item to a new line in the file
    for item in url:
        file.write(item + "\n")
    
```

```python
driver = webdriver.Chrome(executable_path='chromedriver.exe')
url = []
base_url = 'https://www.mayoclinic.org'
for cat in ["cancer", "cardiovascular-diseases", "digestive-diseases", "endocrinology", "neurology-neurosurgery", "obstetrics-gynecology", "ophthalmology", "orthopedic-surgery", "pediatrics", "physical-medicine-rehabilitation", "psychiatry-psychology", "pulmonary-medicine", "transplant-medicine", "trauma", "urology"]:
    for x in range(1, 50):
        u = "https://www.mayoclinic.org/medical-professionals/{}/physicians?lastInitial=&page={}".format(cat, x)
        driver.get(u)
        time.sleep(1)

        soup = BeautifulSoup(driver.page_source, 'html.parser')
        try:
            data = soup.find("div", class_='grid-list col-2-clickable doctort-list content-within')
            if not data.find("li"):
                break
            for y in data.find_all("li"):
                try:
                    url.append(base_url+y.find('a', href=True)['href'])
                except:
                    pass
        except:
            break
```

```python
def get_urls5(urls):
    name = []
    department = []
    specialties = []
    about = []
    education = []
    location = []
    link = []

    for u in urls:
        try:
            r = requests.get(u)
        except:
            continue
        soup = BeautifulSoup(r.content, 'html5lib')
        
        link.append(u)
        name.append(soup.find("h1").text.strip().split(",")[0])

        data = soup.find_all("ul", class_="msk-list msk-list--unordered")
        try:
            temp = ''
            for x in data[1].find_all("li"):
                temp += x.text.replace("\n", "")+' | '
            department.append(temp[:-2])
        except:
            department.append("")
        try:
            temp = ''
            for x in data[2].find_all("li"):
                temp += x.text.replace("\n", "")+' | '
            specialties.append(temp[:-2])
        except:
            specialties.append("")
        try:
            temp = ''
            for x in data[3].find_all("li"):
                temp += x.text.replace("\n", "")+' | '
            education.append(temp[:-2])
        except:
            education.append("")
        try:
            about.append(soup.find("div", class_="msk-markup").find("p").text.strip().replace("\n", "").replace("\xa0", " "))
        except:
            about.append("")
        try:
            location.append(soup.find("div", class_="doctor-contact-location__data-point doctor-contact-location__data__physical").find("div").text.split(",")[0].strip().replace("\n", ""))
        except:
            location.append("")
    d = {"Name":name, "Department Profile":department, "Specialties":specialties, "Education":education, "Location":location, "About":about, "Profile Link":link}
    return d


def get_urls4(urls):
    name = []
    department = []
    specialties = []
    about = []
    education = []
    location = []
    link = []
    for u in urls:
        try:
            r = requests.get(u)
        except:
            continue
        soup = BeautifulSoup(r.content, 'html5lib')
        try:
            bio = soup.find("div", class_="bio")
            name.append(bio.find("h4", class_="physician-profile-title").text.strip())
        except:
            continue
        link.append(u)
        try:
            department.append(bio.find("p", class_="physician-academic-title").text.strip())
        except:
            department.append("")
        try:
            data = soup.find("div", class_="row clearfix ")
            temp = ""
            for x in data.find_all("li"):
                temp += x.text.replace("\n", "")+' | '
            specialties.append(temp[:-2])
        except:
            try:
                specialties.append(bio.find("h4", class_="physician-title").text.strip())
            except:
                specialties.append("")
                
        try:
            data = soup.find("div", class_="noindex accordion-content list-wrapper open")
            temp = ''
            for x in data.find_all("li"):
                temp += x.text.replace("\n", "")+' | '
            education.append(temp[:-2])
        except:
            education.append("")
        try:
            about.append(soup.find("div", class_="bio-txt view-more-section").text.strip().replace("\n", "").replace("\xa0", " "))
        except:
            about.append("")
        try:
            location.append(soup.find("span", class_="subtitle not-mobile").text.strip())
        except:
            location.append("")
    d = {"Name":name, "Department Profile":department, "Specialties":specialties, "Education":education, "Location":location, "About":about, "Profile Link":link}
    return d

def get_urls3(urls):
    name = []
    department = []
    specialties = []
    about = []
    education = []
    location = []
    link = []
    
    for u in urls:
        try:
            r = requests.get(u)
        except:
            continue
        soup = BeautifulSoup(r.content, 'html5lib')
        try:
            name.append(soup.find("div", class_="name").text.strip().split(",")[0])
        except:
            continue
        link.append(u)
        try:
            data = soup.find("ul", class_="titles")
            temp = ''
            for x in data.find_all("li"):
                temp += x.text.replace("\n", "")+' | '
            department.append(temp[:-2])
        except:
            department.append("")
            
        try:
            data = soup.find("div", class_="expertise").text.strip()
            specialties.append(data.split("\n")[-1][:-13].strip())
        except:
            specialties.append("")
        try:
            data = soup.find("div", class_="section education").find_all("ul")
            temp = ''
            for x in data[0].find_all("li"):
                temp += x.text.replace("\n", "")+' | '
            education.append(temp[:-2])
        except:
            education.append("")
        try:
            about.append(soup.find("div", class_="bio biography-text").text.strip().replace("\n", "").replace("\xa0", " "))
        except:
            about.append("")
        try:
            location.append(re.sub(r'\s{1,}', ' ', soup.find("div", class_="address").text.strip().replace("\n", ""))[:-4])
        except:
            location.append("USA")
    d = {"Name":name, "Department Profile":department, "Specialties":specialties, "Education":education, "Location":location, "About":about, "Profile Link":link}
    return d

def get_urls2(urls):
    name = []
    department = []
    specialties = []
    about = []
    education = []
    location = []
    link = []
    for u in urls:
        try:
            r = requests.get(u)
        except:
            continue
            
        soup = BeautifulSoup(r.content, 'html5lib')
        
        try:
            data = soup.find("div", class_="hero-content__description").text.strip().replace("\n", "|")
            if len(data)< 4:
                  continue
            department.append(data)
        except:
            continue
            
        try:
            name.append(soup.find("h1", class_="hero-content__title").text.strip().split(",")[0])
        except:
            continue
            
        link.append(u)
        
        try:
            data = soup.find_all("ul", class_="links-columns")
            temp = ''
            for x in data[1].find_all("li"):
                temp += x.text.replace("\n", "")+' | '
            specialties.append(temp[:-2])
        except:
            specialties.append("")
            
        try:
            data = soup.find("div", class_="profile-detail__block")
            temp = ''
            for x in data.find_all("li"):
                temp += x.text.replace("\n", "")+' | '
            education.append(temp[:-2])
        except:
            education.append("")
            
        try:
            about.append(soup.find("div", class_="read-more__content").find_all("p")[1].text.strip().replace("\n", "").replace("\xa0", " "))
        except:
            about.append("")
            
        try:
            data = soup.find("div", class_="listing-item--location__contact-info").text.replace("\n", " ").replace("\xa0", " ").strip().split(", ")
            data = data[-2].split(" ")[-1].strip()+", "+data[-1].split(" ")[0].strip()
            location.append(data)
        except:
            location.append("USA")
            
    d = {"Name":name, "Department Profile":department, "Specialties":specialties, "Education":education, "Location":location, "About":about, "Profile Link":link}
    return d

def get_urls1(urls):
    name = []
    department = []
    specialties = []
    about = []
    education = []
    location = []
    link = []
    
    for u in urls:
        try:
            r = requests.get(u)
        except:
            continue
        soup = BeautifulSoup(r.content, 'html5lib')
        data = soup.find("h1").text.strip().split(",")[0]
        if data == "Server Error":
            continue
        name.append(data)
        link.append(u)
        
        try:
            data = soup.find("ol", class_="bio-rank")
            temp = ''
            for x in data.find_all("li"):
                temp += x.text.replace("\n", "")+' | '
            department.append(temp[:-2])
        except:
            department.append("")
            
        try:
            data = soup.find("ul", class_="bio-list listGroup")
            temp = ''
            for x in data.find_all("li"):
                temp += x.text.replace("\n", "")+' | '
            specialties.append(temp[:-2])
        except:
            specialties.append("")
            
        try:
            data = soup.find("ol", class_="bio-list listGroup")
            temp = ''
            for x in data.find_all("li"):
                temp += x.text.replace("\n", "")+' | '
            education.append(temp[4:-2])
        except:
            education.append("")
        try:
            about.append(soup.find("div", class_="bio-summary").text.strip().replace("\n", "").replace("\xa0", " "))
        except:
            about.append("")
        try:
            location.append(soup.find("div", class_="bio-location").find("h3").text.strip().replace("\n", ""))
        except:
            location.append("")
    d = {"Name":name, "Department Profile":department, "Specialties":specialties, "Education":education, "Location":location, "About":about, "Profile Link":link}
    return d

def get_urls(urls):
    name = []
    department = []
    specialties = []
    about = []
    education = []
    location = []
    link = []
    
    for u in urls:
        try:
            r = requests.get(u)
        except:
            continue
        soup = BeautifulSoup(r.content, 'html5lib')
        try:
            name.append(soup.find("h1", class_="header--bio__name lede").text.strip().split(",")[0])
        except:
            continue
        
        link.append(u)

        try:
            department.append(soup.find("span", class_="credentials__item").text.strip())
        except:
            department.append("")
        try:
            r1 = requests.get(u+"#education-content")
            soup1 = BeautifulSoup(r1.content, 'html5lib')
            
            data = soup.find("div", class_="main-content l-66-33--1").find_all("p")
            try:
                specialties.append(data[-1].text.strip())
            except:
                specialties.append("")

            try:
                temp = ''
                for x in data[1:-1]:
                    temp += x.find("strong").text.strip()+' | '
                education.append(temp[:-2])
            except:
                education.append("")
        except:
            specialties.append("")
            education.append("")
            
        try:
            about.append(soup.find("div", class_="main-content l-66-33--1 has-side").text.strip().replace("\n", "").replace("\xa0", " "))
        except:
            about.append("")
            
        try:
            data = soup.find("div", class_="bio--locations-specialties__mail-address").text.replace("\n", " ").replace("\xa0", " ").strip().split(", ")
            data = data[-2].split(" ")[-1].strip()+", "+data[-1].split(" ")[0].strip()
            location.append(data)
        except:
            location.append("")
    
    d = {"Name":name, "Department Profile":department, "Specialties":specialties, "Education":education, "Location":location, "About":about, "Profile Link":link}
    return d
```
