## **Converting the simple text into training data format**

```
with open("/content/inputfile.txt") as f:
    contents = f.readlines()
```

```
text_data = "".join(contents)
text_data[:500]
```

```
import re
import pandas as pd
def extract_information(text):
    info = []
    sections = re.split(r'\n\n\n', text)  # Split by double newline to identify sections
    for section in sections:
        d = {}
        lines = section.split('\n')
        for line in lines:
            temp = line.split(': ')
            d[temp[0]] = temp[-1].strip()
        info.append(d)
    return info
data = extract_information(text_data)
```

```
data[:5]
```

```
train = []
i = 0
for person in data[:-800]:
  context = f"{person['NAME']} is doctor living in {person['LOCATION']} and he is working as a {person['DEPARTMENT']} in department. He has specialties on {person['SPECIALTIES']}. If we talking about his educational qualification then he did {person['EDUCATION']}. He is saying about himself that {person['ABOUT']}"
  train.append({"context": context,
   "qas":[{"answers":[{'answer_start': context.find('doctor'), 'text': 'doctor'}],
           "id": str(10000000+i),
           "is_impossible": False,
           "question":f"Who is {person['NAME']}?"
   },{"answers":[{'answer_start': context.find(f"{person['LOCATION']}"), 'text': f"{person['LOCATION']}"}],
           "id": str(10000000+i+1),
           "is_impossible": False,
           "question":f"What is the location address of {person['NAME']}?"
   },{"answers":[{'answer_start': context.find(f"{person['DEPARTMENT']}"), 'text': f"{person['DEPARTMENT']}"}],
           "id": str(10000000+i+2),
           "is_impossible": False,
           "question":f"What is department of {person['NAME']}?"
   },{"answers":[{'answer_start': context.find(f"{person['SPECIALTIES']}"), 'text': f"{person['SPECIALTIES']}"}],
           "id": str(10000000+i+3),
           "is_impossible": False,
           "question":f"What are specialties of {person['NAME']}?"
   },{"answers":[{'answer_start': context.find(f"{person['EDUCATION']}"), 'text': f"{person['EDUCATION']}"}],
           "id": str(10000000+i+4),
           "is_impossible": False,
           "question":f"What are education qualification of {person['NAME']}?"
   }]})
  i += 5
```

## **Model training**

```
import logging
from simpletransformers.question_answering import QuestionAnsweringModel, QuestionAnsweringArgs
```

```
model_type="bert"
model_name= "bert-base-cased"
```

```
### Advanced Methodology
train_args = {
    "reprocess_input_data": True,
    "overwrite_output_dir": True,
    "use_cached_eval_features": True,
    "output_dir": f"outputs/{model_type}",
    "best_model_dir": f"outputs/{model_type}/best_model",
    "evaluate_during_training": True,
    "max_seq_length": 128,
    "num_train_epochs": 5,
    "evaluate_during_training_steps": 1000,
    "save_model_every_epoch": False,
    "save_eval_checkpoints": False,
    "n_best_size":3,
    # "use_early_stopping": True,
    # "early_stopping_metric": "mcc",
    # "n_gpu": 2,
    # "manual_seed": 4,
    # "use_multiprocessing": False,
    "train_batch_size": 128,
    "eval_batch_size": 64,
}
```

```
model = QuestionAnsweringModel(
    model_type,model_name, args=train_args
)
```

```
### Remove output folder
!rm -rf outputs
```

```
# Train the model
model.train_model(train, eval_data=test)
```

```
# Evaluate the model
result, texts = model.eval_model(test)
```
